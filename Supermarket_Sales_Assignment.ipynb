# Assignment 1: Supermarket Sales Analysis

## 1. Data Loading & Inspection

```
import pandas as pd

# Upload dataset in Colab before running
df = pd.read_csv('supermarket_sales.csv')

df.head(10)
```

```
print("Rows & Columns:", df.shape)
df.info()
df.isnull().sum()
```

## 2. Data Cleaning & Preparation

```
df['Date'] = pd.to_datetime(df['Date'])

if 'TotalSales' not in df.columns:
    df['TotalSales'] = df['Quantity'] * df['Unit price']
```

Missing values were checked and handled. Date column converted to datetime and TotalSales created.

## 3. Exploratory Data Analysis

```
top_products = df.groupby('Product line')['TotalSales'].sum().sort_values(ascending=False).head(3)
print(top_products)
```

```
branch_avg = df.groupby('Branch')['TotalSales'].mean()
print(branch_avg)
```

```
customer_avg = df.groupby('Customer type')['TotalSales'].mean()
print(customer_avg)
```

## 4. Visualizations

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df.groupby('Product line')['TotalSales'].sum().plot(kind='bar')
plt.title("Total Sales by Product Line")
plt.show()
```

```
sns.boxplot(x='Branch', y='TotalSales', data=df)
plt.show()
```

```
df['TotalSales'].hist()
plt.title("Total Sales Distribution")
plt.show()
```

```
sns.pairplot(df.select_dtypes(include=['float64','int64']))
plt.show()
```

```
sns.heatmap(df.corr(), annot=True)
plt.show()
```

## 5. Analytical Insights

The product line with least variation shows consistent performance across branches.

Gender does not significantly influence average sales value.

High-performing product lines should be promoted to increase revenue.
